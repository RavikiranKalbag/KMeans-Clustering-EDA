## Elbow Method for finding Optimal number of Clusters

- The Elbow Method is used to find out the optimal number of clusters that can be formed out of the Data. Also, the K value should be such that the overall sum of squared distances from centroids should be minimum.

```python
Sum_of_squared_distances = []
K = range(1,21)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data)
    Sum_of_squared_distances.append(km.inertia_)
plt.plot(K, Sum_of_squared_distances, marker = 'o')
plt.xlabel('K')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.grid()
plt.show()
```

- After K = 4, there is a elbow cut in the figure. So, we can conclude that K = 4 from Elbow Method

## Silhouette Method

- The silhouette Method is also a method to find the optimal number of clusters and interpretation and validation of consistency within clusters of data.

```python
from sklearn.metrics import silhouette_score, silhouette_samples

K = range(2,21)
for k in K:
    km = KMeans (n_clusters=k)
    y_pred = km.fit_predict(data)
    centers = km.cluster_centers_

    score = silhouette_score(data, y_pred, metric='euclidean')
    print ("For k = {}, silhouette score is {}".format(k, score))
```

### Visualization of Silhouette Method

```python
from yellowbrick.cluster import SilhouetteVisualizer

km = KMeans (n_clusters=4)
visualizer = SilhouetteVisualizer(km)

visualizer.fit(data)
visualizer.show() 
```

- After K = 4, there is no significant change in the Silhouette Score. So, from both Elbow & Silhouette Method it is shown that K =4.

# Factor Analysis

```python
from factor_analyzer import FactorAnalyzer
```

```python
data.drop("Cluster", axis = 1, inplace = True)
```

```python
data.columns
```

```python
x =data[data.columns] 

fa = FactorAnalyzer()
fa.fit(x, 10)

#Get Eigen values and plot
ev, v = fa.get_eigenvalues()
ev
plt.plot(range(1,x.shape[1]+1),ev, marker = 'o')
```

- After 3 there is no more significant decrease in the eigen values. So, we consider the number of factors as 3.

## Factor Loadings using Varimax Rotation 

```python
fa = FactorAnalyzer(3, rotation='varimax')
fa.fit(x)
loads = fa.loadings_
print(loads)
```

- Based on the Loadings, The three Factors are:
-
1) Purchases
2) Balance
3) Frequency
