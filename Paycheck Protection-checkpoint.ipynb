## Missing Value Treatment

```python
data.isna().sum()
```

```python
data.drop(['BorrowerName', 'BorrowerAddress'], axis = 1, inplace = True)
```

```python
data['NonProfit'].value_counts()
```

```python
data['NonProfit'].fillna(value = 'N', inplace = True)
data['ForgivenessAmount'] = data['ForgivenessAmount'].fillna(data['ForgivenessAmount'].mean())
```

```python
data['BusinessType'].fillna(data['BusinessType'].mode()[0], inplace = True)
```

```python
data.isna().sum()
```

```python
print(data['DEBT_INTEREST_PROCEED'].value_counts())
print(data['HEALTH_CARE_PROCEED'].value_counts())
print(data['REFINANCE_EIDL_PROCEED'].value_counts())
print(data['RENT_PROCEED'].value_counts())
print(data['PAYROLL_PROCEED'].value_counts())
print(data['MORTGAGE_INTEREST_PROCEED'].value_counts())
print(data['UTILITIES_PROCEED'].value_counts())
```

```python
data['DEBT_INTEREST_PROCEED'].fillna(0.00, inplace = True)
data['HEALTH_CARE_PROCEED'].fillna(0.00, inplace = True)
data['REFINANCE_EIDL_PROCEED'].fillna(0.00, inplace = True)
data['RENT_PROCEED'].fillna(0.00, inplace = True)
data['MORTGAGE_INTEREST_PROCEED'].fillna(0.00, inplace = True)
data['UTILITIES_PROCEED'].fillna(0.00, inplace = True)
data['PAYROLL_PROCEED'].fillna(150000.00, inplace = True)
```

```python
data.isna().sum()
```

```python
data['BorrowerCity'].fillna(data['BorrowerCity'].mode()[0], inplace = True)
data['BorrowerState'].fillna(data['BorrowerState'].mode()[0], inplace = True)
data['BorrowerZip'].fillna(data['BorrowerZip'].mode()[0], inplace = True)
data['ProjectCity'].fillna(data['ProjectCity'].mode()[0], inplace = True)
data['ProjectCountyName'].fillna(data['ProjectCountyName'].mode()[0], inplace = True)
data['ProjectState'].fillna(data['ProjectState'].mode()[0], inplace = True)
data['ProjectZip'].fillna(data['ProjectZip'].mode()[0], inplace = True)
```

```python
data.isna().sum()
```

```python
data.drop(["CD", "NAICSCode"], axis = 1, inplace = True)
```

```python
data.drop(['ForgivenessDate', 'FranchiseName', 'LoanStatusDate'], axis = 1, inplace = True)
```

```python
data.isna().sum()
```

```python
data['JobsReported'].value_counts()
```

```python
data['JobsReported'].fillna(15.0, inplace = True)
data['UndisbursedAmount'].fillna(data['UndisbursedAmount'].mean(), inplace = True)
```

```python
data.isna().sum()
```

- All missing values are being replaced successfully.
